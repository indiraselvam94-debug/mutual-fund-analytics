```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load datasets
nav = pd.read_csv("data/processed/nav_history_clean.csv")
performance = pd.read_csv("data/generated/scheme_performance.csv")

# Convert date column
nav["date"] = pd.to_datetime(nav["date"])

print("NAV Dataset Shape:", nav.shape)
print("Performance Dataset Shape:", performance.shape)

nav.head()
```

```python
nav = nav.sort_values(["scheme_name", "date"])

nav["daily_return"] = nav.groupby("scheme_name")["nav"].pct_change()

nav["daily_return"].describe()
```

```python
# CAGR Calculation

cagr_results = []

for scheme, group in nav.groupby("scheme_name"):
    group = group.sort_values("date")

    end_nav = group.iloc[-1]["nav"]
    end_date = group.iloc[-1]["date"]

    result = {"scheme_name": scheme}

    for years in [1, 3, 5]:
        start_date = end_date - pd.DateOffset(years=years)
        past = group[group["date"] <= start_date]

        if len(past) > 0:
            start_nav = past.iloc[-1]["nav"]
            cagr = (end_nav / start_nav) ** (1 / years) - 1
            result[f"CAGR_{years}Y"] = cagr
        else:
            result[f"CAGR_{years}Y"] = np.nan

    cagr_results.append(result)

cagr_df = pd.DataFrame(cagr_results)

cagr_df
```

```python
# Sharpe Ratio Calculation

risk_free_rate = 0.065  # 6.5% annual
daily_rf = risk_free_rate / 252

sharpe_results = []

for scheme, group in nav.groupby("scheme_name"):

    returns = group["daily_return"].dropna()

    if len(returns) > 0:

        sharpe = ((returns.mean() - daily_rf) / returns.std()) * np.sqrt(252)

        sharpe_results.append({
            "scheme_name": scheme,
            "Sharpe_Ratio": sharpe
        })

sharpe_df = pd.DataFrame(sharpe_results)

sharpe_df = sharpe_df.sort_values(
    "Sharpe_Ratio",
    ascending=False
)

sharpe_df
```

```python
# Sortino Ratio Calculation

risk_free_rate = 0.065
daily_rf = risk_free_rate / 252

sortino_results = []

for scheme, group in nav.groupby("scheme_name"):

    returns = group["daily_return"].dropna()

    downside_returns = returns[returns < daily_rf]

    if len(downside_returns) > 0:

        downside_std = downside_returns.std()

        sortino = ((returns.mean() - daily_rf) / downside_std) * np.sqrt(252)

    else:
        sortino = np.nan

    sortino_results.append({
        "scheme_name": scheme,
        "Sortino_Ratio": sortino
    })

sortino_df = pd.DataFrame(sortino_results)

sortino_df = sortino_df.sort_values(
    "Sortino_Ratio",
    ascending=False
)

sortino_df
```

```python
# Maximum Drawdown Calculation

drawdown_results = []

for scheme, group in nav.groupby("scheme_name"):

    group = group.sort_values("date").copy()

    group["Running_Max"] = group["nav"].cummax()

    group["Drawdown"] = (group["nav"] / group["Running_Max"]) - 1

    max_dd = group["Drawdown"].min()

    worst_row = group.loc[group["Drawdown"].idxmin()]

    drawdown_results.append({
        "scheme_name": scheme,
        "Maximum_Drawdown": max_dd,
        "Worst_Date": worst_row["date"]
    })

drawdown_df = pd.DataFrame(drawdown_results)

drawdown_df = drawdown_df.sort_values(
    "Maximum_Drawdown"
)

drawdown_df
```

```python
# Fund Scorecard (Temporary - Without Alpha)

scorecard = cagr_df.merge(sharpe_df, on="scheme_name")
scorecard = scorecard.merge(drawdown_df, on="scheme_name")

# Ranking
scorecard["Return_Rank"] = scorecard["CAGR_3Y"].rank(ascending=False)
scorecard["Sharpe_Rank"] = scorecard["Sharpe_Ratio"].rank(ascending=False)
scorecard["Drawdown_Rank"] = scorecard["Maximum_Drawdown"].rank(ascending=False)

# Temporary score
scorecard["Fund_Score"] = (
    scorecard["Return_Rank"] * 0.5 +
    scorecard["Sharpe_Rank"] * 0.3 +
    scorecard["Drawdown_Rank"] * 0.2
)

scorecard = scorecard.sort_values("Fund_Score")

scorecard
```
